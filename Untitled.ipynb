```python
save_dir = "images/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
query = "elon musk"
response = requests.get(f"https://www.google.com/search?q={query}sca_esv=561891653&q=elon+musk&tbm=isch&source=lnms&sa=X&ved=2ahUKEwim8Y7vkYmBAxUkSGwGHVQRAXgQ0pQJegQIDhAB&biw=1280&bih=563&dpr=1.5")
```

```python
soup=BeautifulSoup(response.content, "html.parser")
```

```python
 image_tages=soup.find_all("img")
```

```python
del image_tages[0]
```

```python
len(image_tages)
```

```python
image_tages
```

```python
image_data_mongo =[]
for i in image_tages:
    image_url = i['src']
    image_data = requests.get(image_url).content
    mydict = {"index" : image_url , "images" : image_data}
    image_data_mongo.append(mydict)
    with open(os.path.join(save_dir,f"{query}_{image_tages.index(i)}.jpg") , "wb") as f :
        f.write(image_data)
```
